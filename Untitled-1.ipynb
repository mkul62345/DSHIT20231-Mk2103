```python
def getsoup(input_url):
    headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }
    req = requests.get(input_url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    return soup


def saveScrapelist(locsoup):
    #Creates the list
    tbl = locsoup("table" , attrs = {"class":"stats-table player-ratings-table"})[0]
    scrapelist = [t.a.get('href')[15:] for t in tbl.find_all("td", attrs = {"class":"playerCol"})]

    #Saves all to a text file
    f = open("S:/playerseq.txt", "w")
    for plink in scrapelist:
        f.write(plink+"\n")
    f.close()

    return scrapelist


#url = "https://www.hltv.org/stats/players?minMapCount=10"
#url = "https://www.hltv.org/stats/players?startDate=2022-01-01&endDate=2022-12-31&minMapCount=1"
#soup = getsoup(url)
#plinks = saveScrapelist(soup)
plinks = ['11893/zywoo'] #Main scraping is commented out to not rescrape every time we run the notebook since scraping our data took 4-5h

```

The columns that have categorical information in string form we will transform with a label encoder:

The columns that have miscellaneous information will get dropped

```python
def transfer_str_to_numeric_vals(dataset,str_cols,id_cols_to_remove):
    for str_col in str_cols:
        label_encoder = preprocessing.LabelEncoder()
        dataset[str_col] = label_encoder.fit_transform(dataset[str_col])
    dfRet = dataset.drop(columns = id_cols_to_remove).copy()
    return dfRet


#This list will let us understand which player each record relates to, from this point we remember not to reset indexes.
playerLegend = df["nickname"].values.tolist()

#Taking care of none number columns
colstonum = [ 'nationality', 'current_team' ]
colstorem = ['nickname', 'full_name']
dfnums = transfer_str_to_numeric_vals(df, colstonum,colstorem)

dfnums
```

Now that our dataset is clean and free of non-numeric values, we can move on.
 The next step will be splitting the data into the target column (y) and the rest of the dataset (X)
 After that we use the train_test_split() function from sklearn to split the X and y data into random test and train subgroups.
 We will use a test to train ratio of 1:4 (test_size = 0.2)

```python
def split_to_train_and_test(dataset, label_column):

    X = dataset
    y = dataset[label_column].squeeze()
    X.drop(label_column, axis = 1, inplace = True)    
    
    X_train, X_test, y_train, y_test = train_test_split(X,y , random_state = 42 ,test_size = 0.2)
    
    return X_train, X_test, y_train, y_test

category_col_name = 'awards'
X_train, X_test, y_train, y_test = split_to_train_and_test(dfnums, category_col_name)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Now that we have established a baseline of performance of the classifiers, we want to do some hyperparameter tuning in order to improve the model.

Tuning the parameters of the decision tree with the parameter grid:
'min_samples_split' : {1-40}
'max_depth' : None, {1-10}
'min_samples_leaf': {1-20}

```python
def find_best_params_for_dtree(X_train, y_train):
 
    Dtree= tree.DecisionTreeClassifier()

    #Param choices inspired by article : https://towardsdatascience.com/how-to-tune-a-decision-tree-f03721801680
    parameters = {'min_samples_split' : [1,2,3,5,7,10,12,15,20,25,30,35,40], 'max_depth' : [None, 1,2,3,5,7,10], 'min_samples_leaf': [1,2,3,5,7,10,12,15,20]}
    f1scorer = make_scorer(metrics.f1_score)
    
    grid = GridSearchCV(Dtree, parameters, scoring = f1scorer)    
    grid.fit(X_train, y_train)
    
    best_min_split = grid.best_params_['min_samples_split']
    best_min_leaf = grid.best_params_['min_samples_leaf']
    best_max_depth = grid.best_params_['max_depth']

    best_score = grid.best_score_
    
    
    return best_min_split, best_min_leaf , best_max_depth, best_score
```

Here is the block that runs the actual function :

```python
min_split, min_leaf, max_depth, high_score = find_best_params_for_dtree(X_train, y_train)
print(min_split, min_leaf, max_depth, high_score)
```

The best hyperparameters our function found are:
    min_split = 20
    min_leaf = 1
